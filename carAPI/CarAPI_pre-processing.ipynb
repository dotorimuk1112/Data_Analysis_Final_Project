# 1. MYEAR - '0000' 형태로 전처리

```python
def update_year(year):
    if int(year[:2]) > 50:
        return '19' + year[:2]
    else:
        return '20' + year[:2]

df['MYEAR'] = df['MYEAR'].apply(update_year)
```

```python
df.head()
```

# 2. DISP 전처리(VTYPE별 DISP 평균치로 0값 대체)

```python
import numpy as np

# 1. DISP가 0인 값을 NaN으로 바꿉니다.
df['DISP'] = df['DISP'].replace(0, np.nan)

# 2. VTYPE별로 DISP의 평균을 계산합니다.
average_disp_by_vtype = df.groupby('VTYPE')['DISP'].mean()

# 3. 각 VTYPE의 평균 DISP를 사용하여 DISP가 NaN인 행의 값을 채웁니다.
for vtype, avg_disp in average_disp_by_vtype.items():
    df.loc[(df['DISP'].isnull()) & (df['VTYPE'] == vtype), 'DISP'] = avg_disp
```

```python
df['DISP'] = df['DISP'].astype(int)
```

# 3. 팰리세이드 통일하기

```python
# '펠'을 '팰'로 바꿉니다.
df['MNAME'] = df['MNAME'].str.replace('펠', '팰')
```
