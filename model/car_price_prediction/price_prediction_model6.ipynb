```python
from sklearn.cluster import KMeans

df.columns
kmeans = KMeans(n_clusters=3, random_state=42)  # 클러스터 개수는 적절히 조절
kmeans.fit(df)

# 클러스터링 결과 확인
df['cluster'] = kmeans.labels_
print(df['cluster'].value_counts())
```

```python
from xgboost import XGBRegressor

# 각 클러스터별 모델 학습 및 가격 예측
models = {}
for cluster_id in df['cluster'].unique():
    cluster_data = df[df['cluster'] == cluster_id]
    X = cluster_data.drop(['PRICE', 'cluster'], axis=1)  # 입력 변수
    y = cluster_data['PRICE']  # 타겟 변수

    # XGBoost Regressor 모델 생성 및 학습
    model = XGBRegressor()
    model.fit(X, y)
    models[cluster_id] = model

# 각 클러스터별 모델 예측 결과 확인
for cluster_id, model in models.items():
    cluster_data = df[df['cluster'] == cluster_id]
    X_test = cluster_data.drop(['PRICE', 'cluster'], axis=1)  # 테스트 데이터 입력 변수
    y_pred = model.predict(X_test)  # 예측 결과
    print(f"Cluster {cluster_id} 예측 결과:", y_pred)
```

```python
# 각 클러스터별 모델 예측 결과 저장
predictions = pd.DataFrame(columns=['실제값', '예측값'])

for cluster_id, model in models.items():
    cluster_data = df[df['cluster'] == cluster_id]
    X_test = cluster_data.drop(['PRICE', 'cluster'], axis=1)  # 테스트 데이터 입력 변수
    y_test = cluster_data['PRICE']  # 실제값

    # 예측값 계산
    y_pred = model.predict(X_test)

    # 결과를 데이터프레임에 추가
    cluster_predictions = pd.DataFrame({'실제값': y_test, '예측값': y_pred})
    predictions = pd.concat([predictions, cluster_predictions])

# 예측값과 실제값 출력
print(predictions)
```

```python
from sklearn.metrics import mean_squared_error

# 각 클러스터별 모델의 MSE 저장할 딕셔너리 생성
cluster_mses = {}

# 각 클러스터별 모델 예측 결과 저장
for cluster_id, model in models.items():
    cluster_data = df[df['cluster'] == cluster_id]
    X_test = cluster_data.drop(['PRICE', 'cluster'], axis=1)  # 테스트 데이터 입력 변수
    y_test = cluster_data['PRICE']  # 실제값

    # 예측값 계산
    y_pred = model.predict(X_test)

    # 각 클러스터별 MSE 계산
    mse = mean_squared_error(y_test, y_pred)

    # 결과를 딕셔너리에 저장
    cluster_mses[cluster_id] = mse

# 각 클러스터별 MSE 출력
for cluster_id, mse in cluster_mses.items():
    print(f"Cluster {cluster_id} MSE:", mse)
    print(f"Cluster {cluster_id} sqrt MSE:", mse**0.5)
```

```python
for i in df['cluster'].unique():
    print(f'cluster {i}:', df[df['cluster']==i]['PRICE'].mean())
```

```python
for cluster_id, model in models.items():
    # print(cluster_id)
    cluster_data = df[df['cluster'] == cluster_id]
    X_test = cluster_data.drop(['PRICE', 'cluster'], axis=1)  # 테스트 데이터 입력 변수
    y_test = cluster_data['PRICE']  # 실제값

    # 예측값 계산
    y_pred = model.predict(X_test)
    
    print(f"cluster_mses[{cluster_id}]", cluster_mses[cluster_id])
    # 결과를 데이터프레임에 추가
    cluster_predictions = pd.DataFrame({'실제값': y_test,
                                        '예측값': y_pred,
                                        '최소값': y_pred-(cluster_mses[cluster_id]**0.5),
                                        '최대값': y_pred+(cluster_mses[cluster_id]**0.5),
                                        '범위 포함 여부': ((y_pred > y_pred-(cluster_mses[cluster_id]**0.5)) &
                                                          (y_pred < y_pred+(cluster_mses[cluster_id]**0.5))).any()
                                        })


# 예측값과 실제값 출력
print(cluster_predictions)
cluster_predictions.to_csv('pred_result7.csv')
print(cluster_predictions['범위 포함 여부'].value_counts())
```
